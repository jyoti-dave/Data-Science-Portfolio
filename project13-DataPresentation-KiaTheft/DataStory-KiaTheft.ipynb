```python
# Step1: Reading the data from all 6 files and combined them into one.
import warnings
# Suppress all warnings
warnings.filterwarnings("ignore")
import pandas as pd

# # Load raw file with header=None to access first two rows
df_raw = pd.read_excel("Motherboard_VICE_News_Kia_Hyundai_Theft_Data.xlsx", header=None)

# Extract header rows
cities = df_raw.iloc[0].fillna(method='ffill')   # Row 0: Fill down city names
metrics = df_raw.iloc[1]                         # Row 1: Kia/Hyundais, All, Percent

# Combine city + metric into column names
column_names = ['Month'] + [
    f"{str(city).strip().replace(',', '').replace(' ', '_')}_{str(metric).strip().replace(' ', '')}"
    for city, metric in zip(cities[1:], metrics[1:])
]

# Apply new columns
df_clean = df_raw.iloc[2:].copy()
df_clean.columns = column_names

# Parse Month column as datetime
df_clean['Month'] = pd.to_datetime(df_clean['Month'], errors='coerce')

# Reset index
df_clean.reset_index(drop=True, inplace=True)

# Optional: convert numeric columns
for col in df_clean.columns[1:]:
    df_clean[col] = pd.to_numeric(df_clean[col].astype(str).str.replace('%', '').str.strip(), errors='coerce')

# Show preview
print(df_clean.head())
print("Combined CSV Shape:", df_clean.shape)

# Check missing or null values
print("\nMissing values per column:\n")
print(df_clean.isnull().sum())

```

```python


import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as mtick

# Example: Load monthly theft data for Milwaukee
milwaukee_df = pd.read_csv("KiaHyundaiMilwaukeeData.csv")

# Convert to datetime for proper sorting
milwaukee_df['date'] = pd.to_datetime(milwaukee_df['month'] + ' ' + milwaukee_df['year'].astype(str))
milwaukee_df = milwaukee_df.sort_values('date')

# Calculate total thefts
milwaukee_df['Total_Thefts'] = milwaukee_df['countKiaHyundaiThefts'] + milwaukee_df['countOtherThefts']
```

```python
#2. Stacked Bar Chart – Monthly Kia/Hyundai vs Other Thefts


plt.figure(figsize=(12, 6))
plt.bar(milwaukee_df['date'], milwaukee_df['countOtherThefts'], label='Other Cars', color='gray')
plt.bar(milwaukee_df['date'], milwaukee_df['countKiaHyundaiThefts'], bottom=milwaukee_df['countOtherThefts'], label='Kia/Hyundai', color='red')
plt.title('Monthly Car Thefts in Milwaukee (Stacked)')
plt.xlabel('Month')
plt.ylabel('Number of Thefts')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 3. Pie Chart – % of Kia/Hyundai Thefts in 2022 (Top Cities)
# Example city summary
city_summary = pd.read_csv("kiaHyundaiThefts.csv")
top_cities_2022 = city_summary.sort_values(by='percentKiaHyundai', ascending=False).head(5)

labels = top_cities_2022['city']
sizes = top_cities_2022['percentKiaHyundai']

plt.figure(figsize=(7, 7))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("Reds"))
plt.title('Kia/Hyundai Theft Share by City (2022)')
plt.axis('equal')
plt.show()
```

```python
#4. Tree Map – Kia/Hyundai Thefts by City
import squarify

# Summarize Kia/Hyundai thefts by city
city_tree = milwaukee_df.groupby('city').agg({'countKiaHyundaiThefts': 'sum'}).reset_index()
city_tree = city_tree.sort_values(by='countKiaHyundaiThefts', ascending=False).head(15)

plt.figure(figsize=(12, 6))
squarify.plot(sizes=city_tree['countKiaHyundaiThefts'], label=city_tree['city'], alpha=0.8, color=sns.color_palette("Reds"))
plt.title('Kia/Hyundai Thefts by City (Tree Map)')
plt.axis('off')
plt.show()
```

```python
#5. Area Chart – Kia/Hyundai Thefts Over Time
plt.figure(figsize=(12, 6))
plt.fill_between(milwaukee_df['date'], milwaukee_df['countKiaHyundaiThefts'], color='tomato', alpha=0.6)
plt.title('Kia/Hyundai Thefts in Milwaukee Over Time (Area Chart)')
plt.xlabel('Date')
plt.ylabel('Number of Thefts')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#6. Line Chart – % Thefts for Kia/Hyundai
plt.figure(figsize=(12, 6))
plt.plot(milwaukee_df['date'], milwaukee_df['percentKiaHyundai']*100, marker='o', color='red')
plt.title('Percent of Car Thefts that are Kia/Hyundai in Milwaukee')
plt.xlabel('Date')
plt.ylabel('Percent (%)')
plt.xticks(rotation=45)
plt.gca().yaxis.set_major_formatter(mtick.PercentFormatter())
plt.tight_layout()
plt.show()
```

## Ethical Considerations

Data Cleaning & Transformations:

Null or missing months filled where possible; consistent date formatting applied.

Calculated percentages and aggregated theft totals by city/month.

Assumptions & Filtering:

Aggregations excluded cities with >50% missing data. These exclusions will be clearly labeled in the final visuals.

Assumed that all thefts are reported accurately (real-world underreporting not accounted for).

Source & Credibility:

Data sourced from known public safety and crime reporting bodies, assumed accurate.

No sensitive or personal data used.

Presentation Risk:

Avoid fear-mongering—charts will be presented in context and labeled clearly.

Avoid cherry-picking cities—highlighted cities were selected due to complete and consistent data.

Mitigation:

Maintain full transparency on transformations and assumptions in slide notes.

Use accessible design and plain language.
